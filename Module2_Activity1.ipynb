```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# ACTIVITY_1
```

```python

# N = 1
i = 0
X = []
while i < 1000: #number of trials
    x1 = np.random.uniform(0,1,size=1)
    X.append(np.mean(x1))
    i += 1


plt.hist(X)
plt.show
```

```python

# N = 3
i = 0
X = []
while i < 1000: #number of trials
    x1 = np.random.uniform(0,1,size=1)
    x2 = np.random.uniform(0,1,size=1)
    x3 = np.random.uniform(0,1,size=1)
    X.append(np.mean((x1+x2+x3)/3))
    i += 1

plt.hist(X)
plt.show
```

```python
# N = 10
i = 0
X = []
while i < 1000: #number of trials
    x1 = np.random.uniform(0,1,size=20)
    x2 = np.random.uniform(0,1,size=20)
    x3 = np.random.uniform(0,1,size=20)
    x4 = np.random.uniform(0,1,size=20)
    x5 = np.random.uniform(0,1,size=20)
    x6 = np.random.uniform(0,1,size=20)
    x7 = np.random.uniform(0,1,size=20)
    x8 = np.random.uniform(0,1,size=20)
    x9 = np.random.uniform(0,1,size=20)
    x10 = np.random.uniform(0,1,size=20)
    X.append(np.mean((x1+x2+x3+x4+x5+x6+x7+x8+x9+x10)/10))
    i += 1

# Fit a normal distribution to the data
mu, std = norm.fit(X)

# Plot the histogram
plt.hist(X, bins=30, density=True)

# Plot the fitted Gaussian distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)

plt.show()
```
