```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ACTIVITY_1
```

```python

# N = 1
i = 0
X = []
while i < 1000: #number of trials
    x1 = np.random.uniform(0,1,size=1)
    X.append(np.mean(x1))
    i += 1


plt.hist(X)
plt.show
```

```python

# N = 3
i = 0
X = []
while i < 1000: #number of trials
    x1 = np.random.uniform(0,1,size=1)
    x2 = np.random.uniform(0,1,size=1)
    x3 = np.random.uniform(0,1,size=1)
    X.append(np.mean((x1+x2+x3)/3))
    i += 1

plt.hist(X)
plt.show
```
